```python
import plotly.graph_objects as go
import numpy as np
from random import randint
from scipy.special import factorial
from scipy.stats import norm
```

```python
a = 0.5
b = 1

xMean = 4
xSError = 5
```

```python
def yFunc(x, s):
  return (a*x + b + np.random.normal(0, s)*np.exp(- (x - xMean)**2 / (2 * (xSError ** 2))))
```

```python
X = np.random.normal(4, 5, 200)

X.sort()

Yhat = [(a*x + b) for x in X]
Y1 = [yFunc(x, 1) for x in X]
Y2 = [yFunc(x, 5) for x in X]
```

```python
def getMean(RVList):
  return sum(RVList)/len(RVList)
```

```python
def getMSE(Yhat, Y):
  Ydiff = [(yhat - y) for yhat, y in zip(Yhat, Y)]
  YdiffMean = getMean(Ydiff)

  yError = sum([(ydiff - YdiffMean)**2 for ydiff in Ydiff]) / (len(Ydiff))

  return yError
```

```python
print(getMSE(Yhat, Y1)) 
print(getMSE(Yhat, Y2))
```

```python
Y1diff = [(yhat - y1) for yhat, y1 in zip(Yhat, Y1)]
Y2diff = [(yhat - y2) for yhat, y2 in zip(Yhat, Y2)]

fig = go.Figure()
fig.add_trace(go.Scatter(x=X, y=Y1diff, mode='lines', marker_color='red', opacity=0.8))
fig.update_yaxes(range=[-15, 15])
fig.show()
fig.write_image("Y1PlotEst.svg")

fig = go.Figure()
fig.add_trace(go.Scatter(x=X, y=Y2diff, mode='lines', marker_color='blue', opacity=0.8))
fig.update_yaxes(range=[-15, 20])
fig.show()
fig.write_image("Y2PlotEst.svg")
```
